## Cleaning Data

### Year has many Non-year value

```python
backup=car.copy()
```

```python
car=car[car['year'].str.isnumeric()]
```

### Year is boject Change ot integer

```python
car['year']=car['year'].astype(int)
```

### Price has ask for price 

```python
car = car[car['Price']!='Ask For Price']
```

```python
car['Price'] = car['Price'].str.replace(',','').astype(int)
```

### kms_driven has object values with kms at last.

```python
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

### It has non values and two rows have Petrol in thim

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

### fuel_type has non values

```python
car = car[~car['fuel_type'].isna()]
```

### Name and company has spammed data but with the previous cleaning, those rows got emoved


#### Company does not need any cleaning now changing car names. keeping only the first three words

```python
car['name']=car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

### Resetting the index of the final cleaned data

```python
car=car.reset_index(drop=True)
```

### Extracting Training Data

```python
x = car.drop(columns='Price')
y = car['Price'] 
```

```python
x
```

```python
y
```

### Applying Train Test Split


```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import make_column_transformer
```

### Creating an OneHotEncoder object to contain all the possible categories


```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_
```

### Creating a column transformer to transform categorical columns


```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                     remainder='passthrough')
```

### Linear Regression Model

```python
lr=LinearRegression()
```

### Making a pipeline

```python
pipe=make_pipeline(column_trans,lr)
```

### Fitting the model

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

### Checking R2 Score

```python
r2_score(y_test,y_pred)
```

### Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.92 as r2_score

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))    
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

### The best model is found at a certain random state

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
r2_score(y_test,y_pred)
```
